```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style("whitegrid") 
```

```python
data = [['COImini', 130, 881],['COIff', 550, 2288],['COI5P', 658, 2578],['COI3P', 809, 3010],['COI5P3P', 1493, 3657]]

df = pd.DataFrame(data, columns = ['Segment','Base pairs','Unique haplotypes'])

df
```

# Plot data and a linear regression model fit using log(x)

```python
f, ax = plt.subplots(figsize=(8, 4))

ax = sns.regplot(x="Base pairs", y="Unique haplotypes", data=df, logx=True)
sns.set_style("whitegrid")
plt.xlim(0, 1550)
plt.ylim(0, 4200)
f.suptitle('Number of unique haplotypes by segment length', fontsize=14, fontweight='bold')
ax.tick_params(labelsize=12)
ax.set_xlabel("Base pairs of COI",size = 12)
ax.set_ylabel("Unique haplotypes",size = 12)

plt.savefig('Unique_haplotypes.png', dpi=300, bbox_inches="tight")
#plt.savefig('plot.tiff', dpi=300, bbox_inches="tight")
#plt.savefig('plot.pdf', format='pdf', dpi=300, bbox_inches='tight')
#plt.savefig('Bactrocera_dmin_nn_violin.svg', format='svg', dpi=300, bbox_inches='tight')

plt.show()
```
